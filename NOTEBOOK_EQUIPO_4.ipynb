## (1)

A continuaci√≥n, se muestra pivot_table indexada por los Departamentos de inter√©s y a los que se calcula la media de las variables de inter√©s, despu√©s de realizar la agrupaci√≥n, los resultados son:

```python
Departamentos = ['Cauca', 'C√≥rdoba', 'Guain√≠a', 'Guaviare', 'Vaup√©s', 'Vichada', 'Bogot√°, D.C.']
Variables     = ['COBERTURA_NETA', 'TASA_MATRICULACI√ìN_5_16']

data.pivot_table(index = ["DEPARTAMENTO"],
                 values = Variables,
                 aggfunc = {'COBERTURA_NETA' : np.mean,
                            'TASA_MATRICULACI√ìN_5_16' : np.mean,}).loc[Departamentos]
```

para el siguiente gr√°fico, aunque se realiz√≥ el mismo procedimiento anterior, no se us√≥ como funci√≥n ùöóùöô.ùöñùöéùöäùöó  
sino que creamos una funci√≥n para calcular la  ùöôùöäùöõùöùùöíùöåùöíùöôùöäùöåùöíùöòùöó , y es esta la que se grafica. Esto con el √∫nico objetivo de que la gr√°fica resultante.

```python
def Participacion(serie):
    Participacion = serie.sum()/len(serie)*0.01
    return(Participacion)

Punto1 = data.pivot_table(index = ["DEPARTAMENTO"],
                          values = Variables,
                          aggfunc = {'COBERTURA_NETA' : Participacion,
                                     'TASA_MATRICULACI√ìN_5_16' : Participacion}).loc[Departamentos]
```

```python
import plotly.express as px
Figura1 = go.Figure()

MyColors = ["#BF834E", "#F68118", "#F9CA00", "#AEF133", "#19EE9F", "#DE67B4", "#F13057"]
Temp = Punto1.transpose()
for i, color in zip(Temp.columns, MyColors):
    Figura1.add_bar(x = Temp.index, y = Temp[i], marker_color = color, name = i)

Figura1.update_layout(
    # Usando una tema por defecto (https://plotly.com/python/templates/)
    template = "plotly_dark",
    # Personalizaci√≥n de T√≠tulos
    title_text = "PARTICIPACI√ìN OBTENIDA POR DEPARTAMENTO",
    title_font_family = "Open Sans",
    title_font_color = "white",
    title_font_size = 24,
    # Personalizaci√≥n de Ejes
    xaxis = dict(
        tickmode = "array",
        tickvals = [0, 1, 2],
        ticktext = ["Cobertura Neta","Tasa de Matriculaci√≥n"]
    )
)    
Figura1.update_xaxes(
    title_text = "Variable",
    #title_font = dict(size = 20, family = "Balto", color = "#28F618"),
    # Tickes
    tickangle = 0,
    tickfont = dict(size = 16, family = "Overpass", color = "#FFFFFF"),
    # L√≠nea Base
    showline = True, linewidth = 2, linecolor = "#FFFFFF",
)
Figura1.update_yaxes(
    #title_text = "Porcentaje",
    tickformat = "%",
    #title_font = dict(size = 20, family = "Balto", color = "#21ACF3"),
    # Tickes
    tickfont = dict(size = 16, family = "Overpass", color = "#FFFFFF"),
    # L√≠nea Base
    showline = True, linewidth = 2, linecolor = "#FFFFFF", mirror = True
)
Figura1.show()
```
